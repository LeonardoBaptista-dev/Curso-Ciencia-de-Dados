# Podemos fazer filtro usando E / OU , & / |

```python
# Verificando se alguem viajou com a esposa E filho
base[(base.Parch > 1) & (base.SibSp > 1)].head()
```

```python
# O .loc() permite fazer a busca da mesma forma acima, e diferente do exemplo acima ele pode substituir valores
base.loc[(base.Parch >1) | (base.SibSp > 1),['PassengerId','Parch','SibSp']].head()
```
